```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

def fetch_medium_content(username):
    url = f'https://medium.com/@{username}'

    options = Options()
    options.add_argument('--headless')
    options.add_argument('--no-sandbox')
    options.add_argument('--disable-dev-shm-usage')

    driver = webdriver.Chrome(options=options)

    try:
        driver.get(url)

        # Wait for the articles to load
        WebDriverWait(driver, 20).until(EC.presence_of_element_located((By.TAG_NAME, 'article')))

        # Scroll to load more content
        last_height = driver.execute_script("return document.body.scrollHeight")
        while True:
            driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
            time.sleep(2)
            new_height = driver.execute_script("return document.body.scrollHeight")
            if new_height == last_height:
                break
            last_height = new_height

        # Extract article titles and links
        articles = driver.find_elements(By.TAG_NAME, 'article')
        article_data = []
        for article in articles:
            title_element = article.find_element(By.TAG_NAME, 'h2')
            title = title_element.text
            link = title_element.find_element(By.XPATH, './..').get_attribute('href')
            article_data.append({'title': title, 'href': link})

        return article_data

    finally:
        driver.quit()

username = 'roscoe.kerby'
articles = fetch_medium_content(username)
for article in articles:
    print(f"Title: {article['title']}")
    print(f"Link: {article['href']}")
    print()
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import html

def fetch_medium_content(username):
    url = f'https://medium.com/@{username}'

    options = Options()
    options.add_argument('--headless')
    options.add_argument('--no-sandbox')
    options.add_argument('--disable-dev-shm-usage')

    driver = webdriver.Chrome(options=options)

    try:
        driver.get(url)

        # Wait for the articles to load
        WebDriverWait(driver, 20).until(EC.presence_of_element_located((By.TAG_NAME, 'article')))

        # Scroll to load more content
        last_height = driver.execute_script("return document.body.scrollHeight")
        while True:
            driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
            time.sleep(2)
            new_height = driver.execute_script("return document.body.scrollHeight")
            if new_height == last_height:
                break
            last_height = new_height

        # Extract article titles and links
        articles = driver.find_elements(By.TAG_NAME, 'article')
        article_links = []
        for article in articles:
            title_element = article.find_element(By.TAG_NAME, 'h2')
            title = title_element.text
            link = title_element.find_element(By.XPATH, './..').get_attribute('href')
            # Create HTML link wrapped in a paragraph tag with a class
            html_link = f'<p class="medium-article"><a href="{html.escape(link)}" target="_blank" rel="noopener" title="">{html.escape(title)}</a></p>'
            article_links.append(html_link)

        return article_links

    finally:
        driver.quit()

username = 'roscoe.kerby'
article_links = fetch_medium_content(username)

# Print styled HTML
print("""
<style>
    .medium-articles {
        font-family: Arial, sans-serif;
        max-width: 800px;
        margin: 0 auto;
        padding: 20px;
    }
    .medium-article {
        margin-bottom: 15px;
        padding: 10px;
        background-color: #f9f9f9;
        border-radius: 5px;
        transition: background-color 0.3s ease;
    }
    .medium-article:hover {
        background-color: #f0f0f0;
    }
    .medium-article a {
        color: #333;
        text-decoration: none;
        font-weight: bold;
        font-size: 18px;
    }
    .medium-article a:hover {
        color: #0066cc;
    }
</style>
<div class="medium-articles">
""")

for link in article_links:
    print(link)

print("</div>")
```
